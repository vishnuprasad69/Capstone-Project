```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

```

```python
List_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(List_url).text
```

```python
soup = BeautifulSoup(source, 'xml')
```

```python
table=soup.find('table')
column_names = ['Postalcode','Borough','Neighborhood']
df = pd.DataFrame(columns = column_names)
for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        df.loc[len(df)] = row_data

```

```python
df.head()
```

```python
indexNames = df[ df['Borough'] =='Not assigned'].index
df.drop(indexNames , inplace=True)
```

```python
df.head(10)
```

```python
df.loc[df['Neighborhood'] =='Not assigned' , 'Neighborhood'] = df['Borough']
df.head(10)
```

```python
result = df.groupby(['Postalcode','Borough'], sort=False).agg( ', '.join)
df_new=result.reset_index()
df_new.head(15)
```

```python
k=5
toronto_clustering = dfn.drop(['Postalcode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
dfn.insert(0, 'Cluster Labels', kmeans.labels_)

```

```python
import numpy as np
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
```

```python

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(dfn['Latitude'], dfn['Longitude'], dfn['Neighborhood'], dfn['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
